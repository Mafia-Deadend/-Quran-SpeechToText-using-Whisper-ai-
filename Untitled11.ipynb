```python
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torch

# Load the model and processor
processor = WhisperProcessor.from_pretrained("tarteel-ai/whisper-tiny-ar-quran")
model = WhisperForConditionalGeneration.from_pretrained("tarteel-ai/whisper-tiny-ar-quran")


# Load an audio file
import librosa

# Load the audio file
audio_path = '036003.mp3'
audio_data, sampling_rate = librosa.load(audio_path, sr=16000)  # sr=16000 resamples the audio to 16 kHz


print(f'Audio Data Shape: {audio_data.shape}')
print(f'Sampling Rate: {sampling_rate}')


# Assuming you have an audio file loaded as a tensor or a numpy array
audio = torch.tensor(audio_data)  # Replace with your audio data

# Process the audio and generate transcription
inputs = processor(audio, return_tensors="pt", sampling_rate=16000)
generated_ids = model.generate(inputs["input_features"])


# Decode the transcription
transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)
print(transcription)


```
